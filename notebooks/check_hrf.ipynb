```python
from braincoder.hrf import SPMHRFModel, SPMHRFDerivativeModel
import numpy as np
import matplotlib.pyplot as plt

from itertools import product
```

```python
hrf_model = SPMHRFModel(tr=.1)


t = hrf_model.time_stamps

for delay, dispersion in product([4, 5, 6, 8], [.25, .5, 1, 1.5, 2][:1]):
    hrf = hrf_model.get_hrf(hrf_delay=delay, hrf_dispersion=dispersion)
    plt.plot(t, hrf, label=f'delay={delay}, dispersion={dispersion}'    )

    plt.axvline(delay, linestyle='--', color='gray', alpha=0.5)


plt.plot(t, hrf)
```

```python
hrf_model = SPMHRFModel(tr=1.)


t = hrf_model.time_stamps

for delay, dispersion in product([4, 5, 6, 8], [.5, 1, 1.5, 2]):
    hrf = hrf_model.get_hrf(hrf_delay=delay, hrf_dispersion=dispersion)
    plt.plot(t, hrf, label=f'delay={delay}, dispersion={dispersion}'    )

    plt.axvline(delay, linestyle='--', color='gray', alpha=0.5)


plt.plot(t, hrf)

plt.axvline(6.)
```

```python
hrf_model = SPMHRFDerivativeModel(tr=1.)

t = hrf_model.time_stamps

for i, dispersion in enumerate([-1, 0, 1]):
    plt.subplot(1, 3, i + 1)
    for delay_deriv in [-2, 1, 0, 1, 2]:
        hrf = hrf_model.get_hrf(delay_weight=delay_deriv, dispersion_weight=dispersion)
        plt.plot(t, hrf, label=f'delay={delay_deriv}')

plt.gcf().set_size_inches(15, 5)
    # plt.axvline(delay_deriv, linestyle='--', color='gray', alpha=0.5)
# for delay_deriv, dispersion in product([-2, 1, 0, 1, 2], [-1, 0, 1]):
#     hrf = hrf_model.get_hrf(delay_weight=delay_deriv, dispersion_weight=dispersion)
#     plt.plot(t, hrf, label=f'delay={delay_deriv}, dispersion={dispersion}'    )

#     # plt.axvline(delay_deriv, linestyle='--', color='gray', alpha=0.5)
```
